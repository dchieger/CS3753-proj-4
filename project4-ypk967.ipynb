```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
table = soup.find('table', class_='wikitable sortable plainrowheaders')
print(table.prettify())
```

```python

state_names = []
state_abvs = []
state_capitals = []
state_largest_cities = []
state_ratifications = []
state_populations = []
state_areas_mi = []
state_areas_km = []
state_reps = []


for row in table.find_all('tr')[1:]:

    cells = row.find_all('td')


    if len(cells) >= 8:

        state_abvs.append(cells[0].text.strip())
        state_capitals.append(cells[1].text.strip())
        state_largest_cities.append(cells[2].text.strip())
        state_ratifications.append(cells[3].text.strip())
        state_populations.append(cells[4].text.strip())
        state_areas_mi.append(cells[5].text.strip())
        state_areas_km.append(cells[6].text.strip())
        state_reps.append(cells[7].text.strip())

    elif len(cells) == 7:

        state_abvs.append(cells[0].text.strip())
        state_capitals.append(cells[1].text.strip())
        state_largest_cities.append(cells[1].text.strip())
        state_ratifications.append(cells[2].text.strip())
        state_populations.append(cells[3].text.strip())
        state_areas_mi.append(cells[4].text.strip())
        state_areas_km.append(cells[5].text.strip())
        state_reps.append(cells[6].text.strip())

```

```python
#Orignally I pulled the names directly from wikipedia, and was wrapping up the project but luckily realized I was missing a couple states, this function has a map for every states ancrynom so if we are missing data for any states this func will throw us an error
def get_state_full_name(abbreviation):
    states = {
        "AL": "Alabama", "AK": "Alaska", "AZ": "Arizona", "AR": "Arkansas", "CA": "California",
        "CO": "Colorado", "CT": "Connecticut", "DE": "Delaware", "FL": "Florida", "GA": "Georgia",
        "HI": "Hawaii", "ID": "Idaho", "IL": "Illinois", "IN": "Indiana", "IA": "Iowa",
        "KS": "Kansas", "KY": "Kentucky", "LA": "Louisiana", "ME": "Maine", "MD": "Maryland",
        "MA": "Massachusetts", "MI": "Michigan", "MN": "Minnesota", "MS": "Mississippi", "MO": "Missouri",
        "MT": "Montana", "NE": "Nebraska", "NV": "Nevada", "NH": "New Hampshire", "NJ": "New Jersey",
        "NM": "New Mexico", "NY": "New York", "NC": "North Carolina", "ND": "North Dakota", "OH": "Ohio",
        "OK": "Oklahoma", "OR": "Oregon", "PA": "Pennsylvania", "RI": "Rhode Island", "SC": "South Carolina",
        "SD": "South Dakota", "TN": "Tennessee", "TX": "Texas", "UT": "Utah", "VT": "Vermont",
        "VA": "Virginia", "WA": "Washington", "WV": "West Virginia", "WI": "Wisconsin", "WY": "Wyoming"
    }
    return states.get(abbreviation, "MISSING, CORRUPTED DATA")
```

```python

state_full_names = list(map(lambda abv: get_state_full_name(abv), state_abvs))

# Create the DataFrame
state_data = pd.DataFrame({
    'State Abv': state_abvs,
    'State Name': state_full_names,
    'Population': state_populations,
    'Capital': state_capitals,
    'Largest City': state_largest_cities,
    'Ratification': state_ratifications,
    'Area (mi²)': state_areas_mi,
    'Area (km²)': state_areas_km,
    'Representatives': state_reps
})

proper_state_data = pd.DataFrame ({
    'state': state_full_names,
    'population': state_populations
})
```
